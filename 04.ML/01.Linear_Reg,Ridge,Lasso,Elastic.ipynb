```python
# Divide the dataset into independent and dependent

X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.33 , random_state=36)
```

```python
X.shape
```

```python
X_train.shape , X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaler.transform(X_train)
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train , y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
#prediction

y_pred = regression.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
import numpy as np
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test, y_pred)
print(mae)
print(np.sqrt(mse)) #root mean sqrt error
```

```python
##Accuracy r2 and adjucted r2

from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_pred)
```

```python
score
```

```python
#Display adjusted r-squared

1 - (1-score)*(len(y)-1)/(len(y)-X.shape[1]-1)
```
