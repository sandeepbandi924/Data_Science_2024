```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification,make_blobs,make_moons
from sklearn.model_selection import train_test_split
```

```python
X1 , y1 = make_classification(n_samples=1000,n_features=2,n_redundant=0,n_informative=2,
                    n_clusters_per_class=1,class_sep=2.0,random_state=42)
```

```python
X1
```

```python
y1
```

```python
plt.scatter(X1[:,0],X1[:,1],c=y1)
plt.show()
```

```python
X2,y2 =make_blobs(n_samples=1000,centers=2,n_features=2,random_state=42)
```

```python
plt.scatter(X2[:,0],X2[:,1],c=y2)
plt.show()
```

```python
X3,y3 =make_moons(n_samples=1000,noise=0.3,random_state=42)
```

```python
plt.scatter(X3[:,0],X3[:,1],c=y3)
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X1,y1,test_size=0.2,random_state=45)
```

```python
from sklearn.svm import LinearSVC
```

```python
clf = LinearSVC()
clf.fit(x_train,y_train)
```

```python
clf.score(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
x_train3,x_test3,y_train3,y_test3 = train_test_split(X3,y3,test_size=0.2,random_state=45)
```

```python
clf.fit(x_train3,y_train3)
```

```python
clf.score(x_train3,y_train3)
```

```python
y_pred = clf.predict(x_test3)
```

```python
accuracy_score(y_test3,y_pred)
```
